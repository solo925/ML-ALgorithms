Decision Trees are used for both classification and regression tasks. They work by splitting the data into subsets based on the value of input features.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# Example data for a classification task
# X contains two features, y is the binary target variable
X = np.array([[0, 0], [1, 1], [1, 0], [0, 1]])
y = np.array([0, 1, 0, 1])
```

```python
# Initialize and train the model
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
# Make predictions on the training data
y_pred = model.predict(X)
```

```python
# Calculate and print the accuracy
accuracy = accuracy_score(y, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Visualize the decision tree (optional, requires graphviz)
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plot_tree(model, filled=True)
plt.show()
```
